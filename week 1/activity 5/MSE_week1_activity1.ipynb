```python
# Mount Drive and import data
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd
file_path = '/content/drive/MyDrive/load_data/dataset1.csv'
df = pd.read_csv(file_path, header=None)
```

```python
print("Shape (rows, columns):", df.shape)
print("Total elements:", df.size)
```

```python
# Add row labels and column names
df.index = [f'R{i+1}' for i in range(len(df))]
df.columns = [f'C{i+1}' for i in range(len(df.columns))]
```

```python
# Sum and Avg each row and column
df['Row_Sum'] = df.sum(axis=1)
df['Row_Avg'] = df.mean(axis=1)

col_sum = df.sum()
col_avg = df.mean()

col_sum.name = 'Column_Sum'
col_avg.name = 'Column_Avg'


final_df = pd.concat([df, col_sum.to_frame().T, col_avg.to_frame().T])
final_df = final_df.reset_index()
final_df.rename(columns={'index': 'Label'}, inplace=True)
final_df = final_df.round(2)
```

```python
# Save to CSV and display
output_path = '/content/drive/MyDrive/load_data/dataset1_updated.csv'
final_df.to_csv(output_path, index=False)
print(final_df)
```
